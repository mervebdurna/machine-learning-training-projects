```python
import pandas as pd 
```

# Merging DataFrames

```python
staff_df = pd.DataFrame([{'Name': 'Kelly', 'Role' : 'Directo of HR'},
                         {'Name': 'Sally', 'Role' : 'Course liasion'},
                         {'Name': 'James', 'Role' : 'Grader'}])
staff_df.set_index('Name',inplace=True)
staff_df.head()
```

```python
student_df = pd.DataFrame([{'Name': 'James', 'School' : 'Business'},
                         {'Name': 'Mike', 'School' : 'Law'},
                         {'Name': 'Sally', 'School' : 'Engineering'}])
student_df.set_index('Name',inplace=True)
student_df.head()
```

```python
pd.merge(staff_df, student_df, how='outer',right_index=True, left_index= True)
```

```python
pd.merge(staff_df, student_df, how='inner',right_index=True, left_index= True)
```

```python
pd.merge(staff_df, student_df, how='left',right_index=True, left_index= True)
```

```python
staff_df = staff_df.reset_index()
student_df=student_df.reset_index()
```

```python
pd.merge(staff_df, student_df, how='right', on='Name')
```
